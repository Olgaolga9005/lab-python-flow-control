```python
#3. Instead of updating the inventory by subtracting 1 from the quantity of each product, only do it for the products that were ordered (those in "customer_orders").
products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = {}

#Ask the user to input the quantity of each product available in the inventory. Use the product names from the `products` list as keys in the `inventory` dictionary and assign the respective quantities as values.
for product in products:
    quantity = int(input(f"Enter the quantity for {product}: "))
    inventory[product] = quantity #add to the dictionairy: key = assign the value

print("List of inventory: ", inventory)

customer_order = set()

while True:
    p = input("Enter the product you want to order: ")
    customer_order.add(p)

    user_input = input("Do you want to add another product? (yes/no): ").lower()

    if user_input == "yes":
        continue
    else:
        print("All products added to the order.")
        break

print("Customer order:", customer_order)

for product in customer_order:
    if product in inventory:
        inventory[product] -= 1
    else:
        print({product}, "is not in the inventory.")

print("Updated inventory:")
for product, quantity in inventory.items():
    print(product, ":", quantity)

```
